From Random Sample

```python
import numpy as np
import tensorflow as tf
# create a random vector of shape (10,2)
x = np.random.sample((10,2))
# make a dataset from a numpy array
dataset = tf.data.Dataset.from_tensor_slices(x)
for input, output in dataset:
   print (input, output)
```

Features, Labels

```python
features, labels = (np.random.sample((10,1,2)), np.random.sample((10,1,1)))
dataset = tf.data.Dataset.from_tensor_slices((features,labels))
for input, output in dataset:
   print (input, output)
```

From Generator (v1)

```python
# from generator
import numpy as np
import tensorflow.compat.v1 as tf
tf.compat.v1.disable_v2_behavior()

sequence = np.array([[[1]],[[2],[3]],[[3],[4],[5]]])
def generator():
    for el in sequence:
        yield el
dataset = tf.data.Dataset.from_generator(generator,output_types=tf.int64, output_shapes=(tf.TensorShape([None, 1])))
dataset = dataset.batch(1)
iter = dataset.make_initializable_iterator()
el = iter.get_next()
with tf.Session() as sess:
    sess.run(iter.initializer)
    print(sess.run(el))
    print(sess.run(el))
    print(sess.run(el))
```

Model and Dataset

```python
import tensorflow as tf
import numpy as np
train, test = tf.keras.datasets.fashion_mnist.load_data()

images, labels = train
images = images/255.0
labels = labels.astype(np.int32)
```

```python
fmnist_train_ds = tf.data.Dataset.from_tensor_slices((images, labels))
fmnist_train_ds = fmnist_train_ds.shuffle(5000).batch(32)

model = tf.keras.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(10)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
              metrics=['accuracy'])
```

```python
model.fit(fmnist_train_ds, epochs=2)
```

```python
#infinite dataset
#model.fit(fmnist_train_ds.repeat(), epochs=2, steps_per_epoch=20)
```

```python
loss, accuracy = model.evaluate(fmnist_train_ds)
print("Loss :", loss)
print("Accuracy :", accuracy)
```
