```python
import tensorflow as tf
import numpy as np

def vectorize(vector_like):
    return tf.convert_to_tensor(vector_like, dtype=tf.float32)
    #return tf.convert_to_tensor(vector_like)

def batchify(vector):
    '''Make a batch out of a single example'''
    return vectorize([vector])

data = [(batchify([0]), batchify([0, 0])), (batchify([1]), batchify([0, 0])), (batchify([2]), batchify([0, 0]))]

num_hidden = 2
num_classes = 2
opt = tf.keras.optimizers.SGD(learning_rate=0.1)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(num_hidden, activation='relu'))
model.add(tf.keras.layers.Dense(num_classes, activation='sigmoid'))

loss_fn = lambda: tf.keras.losses.mse(model(input), output)
var_list_fn = lambda: model.trainable_weights

for input, output in data:
  opt.minimize(loss_fn, var_list_fn)

model.summary()
print("layer[0]:")
print(model.layers[0].trainable_weights)
print("layer[1]:")
print(model.layers[1].trainable_weights)
```
