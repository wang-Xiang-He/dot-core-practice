```
import tensorflow as tf
```

```
# 載入資料
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print("x_train.shape:", x_train.shape)
print("y_train.shape:", y_train.shape)
y_train[:5,]
```

```
# 建立模型
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

# 1. SGD

```
# 編譯模型
model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
# 訓練模型
r = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10)
```

```
# 評估模型
SGD = model.evaluate(x_test, y_test)
print(SGD)
```

# 2. ADAM

```
# 編譯模型
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
# 訓練模型
r1 = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10)
```

```
# 評估模型
ADAM = model.evaluate(x_test, y_test)
print(ADAM)
```

# 3. RMSProp

```
# 編譯模型
model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
r2 = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10)
```

```
# 評估模型
RMSProp = model.evaluate(x_test, y_test)
print(RMSProp)
```

```
# 繪出每迭代的損失
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = [21, 4]

plt.subplot(131)
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.title("每迭代的損失_SGD");

plt.subplot(132)
plt.plot(r1.history['loss'], label='loss')
plt.plot(r1.history['val_loss'], label='val_loss')
plt.legend()
plt.title("每迭代的損失_ADAM");

plt.subplot(133)
plt.plot(r2.history['loss'], label='loss')
plt.plot(r2.history['val_loss'], label='val_loss')
plt.legend()
plt.title("每迭代的損失_RMSProp");
```

```
# 繪出每迭代的精準度
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = [21, 4]

plt.subplot(131)
plt.plot(r.history['accuracy'], label='acc')
plt.plot(r.history['val_accuracy'], label='val_acc')
plt.legend()
plt.title("每迭代的精準度_SGD");

plt.subplot(132)
plt.plot(r1.history['accuracy'], label='acc')
plt.plot(r1.history['val_accuracy'], label='val_acc')
plt.legend()
plt.title("每迭代的精準度_ADAM");

plt.subplot(133)
plt.plot(r2.history['accuracy'], label='acc')
plt.plot(r2.history['val_accuracy'], label='val_acc')
plt.legend()
plt.title("每迭代的精準度_RMSProp");
```
