我們以最簡單的 損失函數:l=w^2 來展示 Back Propagation 的 Gradient Descent

一階導數:dl/dw=2*w

此例題也不用tensorflow

```
import numpy as np
import matplotlib.pyplot as plt

# 目標函數:l=w^2
def func(w): return np.square(w)

# 目標函數一階導數:dl/dw=2*w
def dfunc(w): return 2 * w

def GD(x_start, df, epochs, lr):    
    """  梯度下降法。給定起始點與目標函數的一階導函數，求在epochs次反覆運算中w的更新值
        :param w_start: w的起始點    
        :param df: 目標函數的一階導函數    
        :param epochs: 反覆運算週期    
        :param lr: 學習率    
        :return: w在每次反覆運算後的位置（包括起始點），長度為epochs+1    
     """    
    ws = np.zeros(epochs+1)    
    w = w_start    
    ws[0] = w    
    for i in range(epochs):         
        dw = df(w)        
        # v表示w要改變的幅度        
        v = - dw * lr        
        w += v        
        ws[i+1] = w    
    return ws
```

```
# Main
# 起始權重
w_start = 5    
# 執行週期數
epochs = 15 
# 學習率   
lr = 0.3   
# 梯度下降法
w = GD(w_start, dfunc, epochs, lr=lr) 
print (w)
# 輸出：從5開始逼近0, 共16個
```

繪出梯度下降圖

```
color = 'r'      
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(w, func(w), c=color, label='lr={}'.format(lr))    
plt.scatter(w, func(w), c=color, label='lr={}'.format(lr) )    
plt.legend()

plt.show()
```
